# Mutual Information

```python
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features = 'auto',n_neighbors=10)
    
    mi_scores = pd.Series(mi_scores,
                          name="MI Scores",
#                           index=['GFD-30', 'GFD-40', 'GFD-50', 'GFD-70', 'GFD-100', 'GFD-140', 'GFD-200', 'GFD-270', 'Pan', 'Moisture', 'Active Clay', 'Clay Content', 'AFS GFN']
                          index=['GFD-30', 'GFD-40', 'GFD-50', 'GFD-70', 'GFD-100', 'GFD-140', 'GFD-200', 'GFD-270', 'Pan', 'Active Clay', 'Clay Content', 'AFS GFN']
                         )
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.xlabel('Mutual Information', fontsize = 13)
```
